# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('ca_takehome_assessment.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
grouped_user_type = df.groupby('user_type').size().reset_index(name='ticket_count')
```

```python
print(grouped_user_type)
```

```python
#making all column names uppercase
df.columns = [col.upper() for col in df.columns]
```

```python
df
```

```python
#changing to datetime

df['CREATED_DAY'] = pd.to_datetime(df['CREATED_DAY'], unit='ms', errors='coerce')
```

```python
df['CLOSED_DAY'] = pd.to_datetime(df['CLOSED_DAY'], unit='ms', errors='coerce')
```

```python
df
```

```python
#dropping rows with important info missing
df.dropna(subset=['USER_TYPE', 'COUNTRY', 'ISSUE_TYPE_CATEGORY'], inplace=True)
```

```python
df
```

```python
df = df[
    (df['CALL_LENGTH_MIN'].isnull()) | ((df['CALL_LENGTH_MIN'] >= 0) & (df['CALL_LENGTH_MIN'] < 1440))
]
df = df[
    (df['CHAT_DURATION_MIN'].isnull()) | ((df['CHAT_DURATION_MIN'] >= 0) & (df['CHAT_DURATION_MIN'] < 1440))
]
df = df[
    (df['FULL_RESOLUTION_TIME'].isnull()) | ((df['FULL_RESOLUTION_TIME'] >= 0) & (df['FULL_RESOLUTION_TIME'] < 1440))
]
```

```python
df
```

```python
#exporting clean data into csv file
df.to_csv('cleaned_df.csv', index=False)
```
